```python
import matplotlib.pyplot as plt
import pandas as pd
import re
import numpy as np

data = pd.read_csv('write.csv')
```

```python
def basicRecommendation(idVisit, column = 'team'):
    tmpData = data[data.actionCategory.isin(['WebNei clicked'])]
    user = tmpData[tmpData.idVisit == idVisit].userName.tail(1).values[0]
    team = tmpData[tmpData.userName == user].sort_values('idVisit').tail(1)[column]
    mostPopular = tmpData[tmpData[column]==team.values[0]].groupby('actionName').size().sort_values(ascending=False)
    watched = tmpData[tmpData['userName']==user]['actionName'].unique()
    s = set(watched)
    result = [(x, mostPopular[x]) for x in mostPopular.index.values if x not in s]
    return pd.DataFrame(result, columns=['name','weight'])
```
