```python
import pandas
import numpy as np
import math
import functools
import bokeh
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

data = pandas.read_csv('C:\\Users\\Robert\\Desktop\\write.csv')
data
```

```python
data["date"]=list(map(lambda x:pandas.to_datetime(x), data.serverTime))
```

```python
def plotAmountPerMonth(columnName, data):
    amountPerMonth = data.groupby([data.date.dt.year, data.date.dt.month])[columnName].nunique()
    years1 = list(map(lambda x: amountPerMonth.index.levels[0][x],amountPerMonth.index.labels[0]))
    months1 = list(map(lambda x: amountPerMonth.index.levels[1][x],amountPerMonth.index.labels[1]))
    date1=list(map(lambda x: str(x[0])+"."+str(x[1]),np.transpose([years1,months1])))
    p = figure(title="simple line example", x_axis_label='miesiÄ…c', y_axis_label='iloÅ›Ä‡ w miesiÄ…cu', x_range=date1)
    p.xaxis.major_label_orientation = math.pi/5
    s = data.groupby("actionCategory").size().sort_values()
    id= [x for x in range (len(amountPerMonth.values))]
    #p.line(date1,amountPerMonth.values, legend="Temp.", line_width=2)
    p.vbar(date1, legend="Temp.", line_width=2,top=amountPerMonth.values,width=0.9)
    show(p)
    output_notebook() 
```

```python
data2 = data[data.actionCategory == "Release technology"]
plotAmountPerMonth("idVisit", data2)
```

```python
data2 = data[data.actionCategory == "Release type"]
plotAmountPerMonth("idVisit", data2)
```
