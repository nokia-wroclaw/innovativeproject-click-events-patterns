```python
from bokeh.core.properties import value
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.io import output_notebook
output_notebook()

fruits = ['Apples', 'Pears', 'Nectarines', 'Plums', 'Grapes', 'Strawberries']
years = ["2015", "2016", "2017"]
colors = ["#c9d9d3", "#718dbf", "#e84d60"]

data = {'fruits' : fruits,
        '2015'   : [2, 1, 4, 3, 2, 4],
        '2016'   : [5, 3, 4, 2, 4, 6],
        '2017'   : [3, 2, 4, 4, 5, 3]}

source = ColumnDataSource(data=data)

p = figure(x_range=fruits, plot_height=250, title="Fruit Counts by Year")

p.vbar_stack(years, x='fruits', width=0.9, color=colors, source=source,
             legend=[value(x) for x in years])

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```

```python
import pandas
import numpy as np
import math
import functools
import bokeh
from bokeh.core.properties import value
from bokeh.io import show, output_file
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.io import output_notebook
output_notebook()
```

```python
data = pandas.read_csv('C:\\Users\\Robert\\Desktop\\write.csv')
```

```python
def getData(data, col1, col2):
    xAxis = []
    yAxis = []
    fields = {}
    g = data.groupby([col2, col1]).size()
    g1=g.unstack()
    g1=g1.fillna(0.0)
    for x in g1:
        yAxis.append(x)
        col = []
        for y_name, y in g1[x].iteritems():
            col.append(y)
            #xAxis.append(y_name)
        fields[x]=col
    xAxis=list(g1.index.values)
    obj = {"xAxis":xAxis, "yAxis":yAxis, "fields":fields}
    return obj
```

```python
obj = getData(data,"actionCategory","department")
fields = obj["fields"]
xAxis = obj["xAxis"]
yAxis = obj["yAxis"]
colors = list(map(lambda x: "#e84d60",yAxis))
```

```python
fields['___x']=xAxis
source = ColumnDataSource(data=fields)

p = figure(x_range=xAxis, plot_height=250, title="Fruit Counts by Year")

p.vbar_stack(stackers=yAxis, x='___x', width=0.9, color=colors, source = source)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
