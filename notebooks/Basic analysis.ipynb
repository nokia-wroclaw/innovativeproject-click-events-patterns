```python
print("Występuje "+str(data.actionCategory.nunique())+" różnych wartości w kolumnie actionCategory.")
print("Wartości te, wraz z ilością wystąpień to: ")
#print(data.groupby("actionCategory").size().sort_values(ascending=False))
print(data.groupby("actionCategory").agg({"idVisit":["count"],"timeSpent":["sum"]}).sort_values(by=[("idVisit","count")],ascending=False)[0:30])
show(Plots.plot(data.groupby("actionCategory").size().sort_values(ascending=False)[0:20]))
```

```python
print("Występuje "+str(data.actionName.nunique())+" różnych wartości w kolumnie actionName.")
print("Najczęściej występujące wartości, wraz z ilością wystąpień to: ")
#print(data.groupby("actionName").size().sort_values(ascending=False)[0:20])
print(data.groupby("actionName").agg({"idVisit":["count"],"timeSpent":["sum"]}).sort_values(by=[("idVisit","count")],ascending=False)[0:30])
show(Plots.plot(data.groupby("actionName").size().sort_values(ascending=False)[0:20]))
print("Różnym actionCategory odpowiadają zwykle różne actionName")
```

```python
print("Występuje "+str(data.userName.nunique())+" różnych wartości w kolumnie userName.")
print("Najczęściej występujące wartości, wraz z ilością wystąpień to: ")
#print(data.groupby("userName").size().sort_values(ascending=False)[0:20])
print(data.groupby("userName").agg({"idVisit":["count"],"timeSpent":["sum"]}).sort_values(by=[("idVisit","count")],ascending=False)[0:30])
__p=Plots.plot(data.groupby("userName").size().sort_values(ascending=False))
__p.xaxis.visible = False
show(__p)
print("Nazwy użtykowników przed dostarczeniem nam zostały zanonimizowane i zastąpione nazwami gier")
```

```python
print("Występuje "+str(data.department.nunique())+" różnych wartości w kolumnie department.")
print("Wartości department, wraz z ilością wystąpień to: ")
#print(data.groupby("department").size().sort_values(ascending=False))
print(data[(data.department!="N/a")&(data.department!="undefined")].groupby("department").agg({"idVisit":["count"],"timeSpent":["sum"]}).sort_values(by=[("idVisit","count")],ascending=False)[0:30])
show(Plots.plot(data[(data.department!="N/a")&(data.department!="undefined")].groupby("department").size().sort_values(ascending=False)[0:20]))
```

```python
print("Występuje "+str(data.team.nunique())+" różnych wartości w kolumnie team.")
print("Najczęściej występujące wartości, wraz z ilością wystąpień to: ")
#print(data.groupby("team").size().sort_values(ascending=False)[0:20])
print(data[data.team!="undefined"].groupby("team").agg({"idVisit":["count"],"timeSpent":["sum"]}).sort_values(by=[("idVisit","count")],ascending=False)[0:30])
#data.groupby("team").department.nunique().sort_values(ascending=False)[0:10]
show(Plots.plot(data[data.team!="undefined"].groupby("team").size().sort_values(ascending=False)[0:20]))
print("Zwykle każdy team należy do tylko jednego departamentu")
print("Występują pojedyńcze wyjątki, gdzie drugi department danego teamu może być 'undefined', lub tytuł pochodzący z innego departmanetu zawiera człon 'ext', lub 'external'.")
```
