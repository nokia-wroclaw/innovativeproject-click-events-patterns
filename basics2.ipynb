```python
def basicRecommendation(idVisit, column):
    user = data[data.idVisit == idVisit].userName.tail(1)
    team = data[data.userName == user].sort_values('idVisit').tail(1)[column]
    mostPopular = data[data[column]==team.values[0]].groupby('actionName').size().sort_values()
    watched = data[data['userName']==user]['actionName'].unique()
    s = set(watched)
    result = [x for x in mostPopular.index.values if x not in s]
    return result
```
