```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('write.csv')
```

```python
mostClickerDepartment = data.groupby("userName").size().sort_values().tail(10)
id = [x for x in range (len(mostClickerDepartment))]
plt.barh(y=id,width=mostClickerDepartment.tolist())
plt.yticks(id,mostClickerDepartment.axes[0].values)
plt.title('Actions by users')
plt.show()
```

```python
mostClickerCountry = data.groupby("country").size().sort_values().tail(10)
id = [x for x in range (len(mostClickerCountry))]
plt.barh(y=id,width=mostClickerCountry.tolist())
plt.yticks(id,mostClickerCountry.axes[0].values)
plt.title('Actions by countries')
plt.show()
```

```python
def histByUniqueVists( column, title ):
    mostUniqueVisits = data[['idVisit', column]].drop_duplicates().groupby(column).size().sort_values().tail(10)
    id = [x for x in range (len(mostUniqueVisits))]
    plt.barh(y=id,width=mostUniqueVisits.tolist())
    plt.yticks(id,mostUniqueVisits.axes[0].values)
    plt.title(title)
    plt.show()
```

```python
histByUniqueVists('department', 'Unique vists by department')
```

```python
histByUniqueVists('userName', 'Unique vists by users')
```

```python
histByUniqueVists('country', 'Unique vists by country')
```
